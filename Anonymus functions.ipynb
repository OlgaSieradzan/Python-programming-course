# Functional Programming with Anonymous Functions in Python

## Lambda Functions (Anonymous Functions)
- **Syntax**: `lambda arguments: expression`
- **Characteristics**:
  - No name (anonymous)
  - Can take any number of arguments but only one expression
  - Automatically returns the result of the expression
  - Often used for short, simple operations

### Example 1: Basic Lambda
```python
square = lambda x: x ** 2
print(square(5))  # Output: 25
```

### Example 2: Multiple Arguments
```python
add = lambda a, b: a + b
print(add(3, 7))  # Output: 10
```

### Comon Use Cases

1. With map()

```python
numbers = [1, 2, 3, 4]
squared = list(map(lambda x: x**2, numbers))
# Result: [1, 4, 9, 16]
```

2. With filter()

```python
numbers = [1, 2, 3, 4, 5, 6]
even = list(filter(lambda x: x % 2 == 0, numbers))
# Result: [2, 4, 6]
```

3. With soreted()

```python
people = [{'name': 'Alice', 'age': 25}, {'name': 'Bob', 'age': 20}]
sorted_people = sorted(people, key=lambda x: x['age'])
# Result: [{'name': 'Bob', 'age': 20}, {'name': 'Alice', 'age': 25}]
```

4. With reduce

```python
from functools import reduce
nums = [1, 2, 3, 4, 5]
print(reduce(lambda a, b: a + b, nums))
```
